```python
def celsius_converter(celsius):
    return (celsius * 9/5) + 32

def fahrenheit_converter(fahrenheit):
    return (fahrenheit - 32) * 5/9

def temperature_converter():
    temperature = float(input("\nType a temperature: "))
    
    # Loop to keep asking until a valid selection is made
    while True:
        conversion_type = input("\nWhat do you want to convert:\n1. Celsius to Fahrenheit\n2. Fahrenheit to Celsius\n")
        if conversion_type == "1":
            result = celsius_converter(temperature)
            print(f"{temperature}째C is equal to {result:.2f}째F")
            break
        elif conversion_type == "2":
            result = fahrenheit_converter(temperature)
            print(f"{temperature}째F is equal to {result:.2f}째C")
            break
        else:
            print("\nInvalid selection. Please choose 1 or 2.")
            
temperature_converter()
```

```python
def Voltage_Calculate(current, resistance):
    return current * resistance

def Current_Calculate(voltage, resistance):
    return voltage / resistance

def Resistance_Calculate(voltage, current):
    return voltage / current

def ohmsLaw_Calculator():
    print("\nWhat do you want to calculate?")
    print("1. Voltage (V)")
    print("2. Current (I)")
    print("3. Resistance (R)")

    while True:
        choice = input("Choose 1, 2, or 3: ")

        if choice == "1":
            try:
                current = float(input("Enter the current (I) in amperes: "))
                resistance = float(input("Enter the resistance (R) in ohms: "))
                voltage = Voltage_Calculate(current, resistance)
                print(f"The voltage is {voltage} volts")
                break
            except ValueError:
                print("Invalid input! Please enter valid number.")
        elif choice == "2":
            try:
                voltage = float(input("Enter the voltage (V) in volts: "))
                resistance = float(input("Enter the resistance (R) in ohms: "))
                if resistance == 0:
                    print("Error: Resistance cannot be zero. Division by zero is not allowed.")
                else:
                    current = Current_Calculate(voltage, resistance)
                    print(f"The current is {current} AMPERES")
                    break
            except ValueError:
                print("Invalid input! Please enter valid number.")
        elif choice == "3":
            try:
                voltage = float(input("Enter the voltage (V) in volts: "))
                current = float(input("Enter the current (I) in amperes: "))
                if current == 0:
                    print("Error: Current cannot be zero. Division by zero is not allowed.")
                else:
                    resistance = Resistance_Calculate(voltage, current)
                    print(f"The resistance is {resistance} ohms")
                    break
            except ValueError:
                print("Invalid input! Please enter valid number.")
        else:
            print("Invalid! Please select between 1, 2, or 3.")

ohmsLaw_Calculator()
```

```python
def print_diamond(n):
    if n % 2 == 0 or n <= 0:
        return "Please input an odd integer."

    for i in range(n // 2 + 1):
        print(' ' * (n // 2 - i) + '*' * (2 * i + 1))
    
    for i in range(n // 2 - 1, -1, -1):
        print(' ' * (n // 2 - i) + '*' * (2 * i + 1))

print_diamond(5)
```
